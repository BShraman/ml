# Data Cleansing

```python
import pandas as pd
import numpy as np
import xlrd

pd.set_option('display.max_columns',None)
players = pd.read_csv("data.csv", delimiter=',',header=0)

players.head()
```

```python
df = players
print(df.columns)
```

```python
# Dropping attributes that are adding no value to dataset
dropList = ['Unnamed: 0','ID','Name','Photo','Flag','Overall','Jersey Number','Club Logo','Body Type', 'Real Face','Joined',
            'Loaned From','Release Clause','Wage']

df.drop(dropList, axis=1, inplace=True)
```

```python
df.shape
```

```python
# Finding the Null value 
df.isnull().sum().sort_values(ascending=False)
```

```python
# Dropping all the columns with Null value
nullList = [index for index, row in df.iterrows() if row.isnull().any()]
print(len(np.unique(nullList)))
```

```python
# Dropping around 12% of Total Data
df.drop(nullList,inplace=True)

df.shape
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Data Analysis for dependent varaible. Since player values are more then 0 
rows_with_zero = df.index[df['Value'] == 0.0].tolist()

print(len(rows_with_zero))
```

```python
# Contract Valid Until, since data are incos
from os import error
from datetime import datetime
import numpy as np

new_CVU = []

for each in df["Contract Valid Until"]:
  if len(each) == 9 :
    data = '20' + each[7:9]
    new_CVU.append(data)
  elif len(each) == 8:
    data = '20' + each[6:8]
    new_CVU.append(data)
  elif len(each) >= 10:
    data = each.split(",")
    new_CVU.append(data[1])
  else:
    data = each
    new_CVU.append(each)
  
df["NRM_CVU"] = new_CVU
```

```python
# Converting K to Mill to have same scale for dependent variable
import decimal

dfValue = df['Value'].str.replace('[€]','')

new_Value = []
for each in dfValue:
  if 'K' in each:
    data = str(round(decimal.Decimal(int(each.replace('K',''))/1000000),6))
    new_Value.append(data)
  else:
    data = each.replace("M",'')
    new_Value.append(data)

df["NRM_Value_Mil"] = new_Value

#Converting object into Float and raise raise exceptions for invalid
df["NRM_Value_Mil"] = df["NRM_Value_Mil"].astype(float, errors = 'raise')
```

```python
# Removing € from row
import decimal
'''
dfWage = df['Wage'].str.replace('[€K]','')

df["NRM_Wage_K"] = dfWage
'''
```

```python
# Converting Height to CM 

dfHeight = df['Height'].str.replace("'",'.')

#Replace Null value with "NA"
dfHeight.fillna("NA", inplace = True)

# 12 incheas to 1 Foot
inch_to_foot = 12

NRM_Height = []
for x in dfHeight:
    data = inch_to_foot * float(x)
    NRM_Height.append(round(data,2))

df["NRM_Height"] = NRM_Height

# Remove Scale from Weight
dfWeight = df['Weight'].str.replace("lbs",'')

df["NRM_Weight_lb"] = dfWeight.astype(float, errors = 'raise')

df["NRM_WorkRate"] = df["Work Rate"].str.replace("/ ",'-')

df.rename(columns={"'Preferred Foot'":'Preferred_Foot'}, inplace=True)
```

```python
import xlrd
from pandas import *
import json
import re 

#SheetName in Supporting Data 
#sheetName = ['NationalityContinent','League','Position']
sheetName = ['NationalityContinent','League']

for sheet in sheetName:
  #Converting excel into json 
  nc_df = pandas.read_excel('SupportingData.xlsx', sheet_name=sheet,header=0)

  #Converting into json
  json_dump =nc_df.to_json(orient='records')

  data = np.array(json.loads(json_dump))
  
  if sheet == 'NationalityContinent':
    dfSheet = df['Nationality'].str.replace('[^A-Za-z0-9�?]+ ', '')
  elif sheet == 'League':
    dfSheet = df['Club'].str.replace('[^A-Za-z0-9�?]+ ', '')
  elif sheet == 'Position':
    dfSheet = df['Position'].str.replace('[^A-Za-z0-9�?]+ ', '')

  NRM_LIST = []
  for each in dfSheet:
    for iteam in data:
      key_list = list(iteam.keys())
      actual_value = key_list[0]
      NRM_value = key_list[1]
      if each == (re.sub('[^A-Za-z0-9�?]+ ', '', iteam[actual_value])):
        NRM_LIST.append(iteam[NRM_value])
      else:
        NRM_LIST.append(iteam[NRM_value])
  
  df["NRM_"+sheet] = Series(NRM_LIST)
```

```python
def getPosition(df):
    if ((df['Position'] == 'RB') | (df['Position'] == 'LB') | (df['Position'] == 'CB') | (df['Position'] == 'LCB') | (df['Position'] == 'RCB') | (df['Position'] == 'RWB') | (df['Position'] == 'LWB') ):
        return 'DF'
    elif ((df['Position'] == 'LDM') | (df['Position'] == 'CDM') | (df['Position'] == 'RDM')):
        return 'DM'
    elif ((df['Position'] == 'LM') | (df['Position'] == 'LCM') | (df['Position'] == 'CM') | (df['Position'] == 'RCM') | (df['Position'] == 'RM')):
        return 'MF'
    elif ((df['Position'] == 'LAM') | (df['Position'] == 'CAM') | (df['Position'] == 'RAM') | (df['Position'] == 'LW') | (df['Position'] == 'RW')):
        return 'AM'
    elif ((df['Position'] == 'RS') | (df['Position'] == 'ST') | (df['Position'] == 'LS') | (df['Position'] == 'CF') | (df['Position'] == 'LF') | (df['Position'] == 'RF')):
        return 'ST'
    else:
        return df.Position
```

```python
df['Player_Position'] = df.apply(getPosition,axis = 1)
```

```python
df.shape
```

```python
# Converting Preferred Foot into Binary indicator
def rightFoot(df):
    if (df['Preferred Foot'] == 'Right'):
        return 1
    else:
        return 0

df['Right_Foot'] = df.apply(rightFoot, axis=1)
```

```python
# Dropping Redundant features
dropFeatureList = ['Preferred Foot','Position','Contract Valid Until','Value','Work Rate','LS', 'ST', 'RS',
                   'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 
                   'LB', 'LCB', 'CB', 'RCB', 'RB','Club','Nationality','Height','Work Rate','Preferred Foot','Weight']

df = df.drop((dropFeatureList) , axis = 1)
```

```python
# ===================   Change Data Types  ===================#

fivestar = ['International Reputation', 'Weak Foot', 'Skill Moves']

for i in fivestar:
  
  df[i]=df[i].values.astype('float64')
  df[i]=df[i].values.astype('str')

#df['NRM_Wage_K']=df['NRM_Wage_K'].values.astype('int64')
```

```python
# ===================   Standardization ===================#
#import category_encoders as ce
from sklearn import preprocessing

sdvDF = df

z_dataType = ['int64','float64']
dv_dataType = ['object']

# Dependent Varaibles
y_list = ['Player_Position','NRM_Value_Mil']

features = sdvDF.columns
dv_list = []
z_list = []

# ============= Preprocessing numerical data ================#
for feature in features:
  if sdvDF.dtypes[feature] in z_dataType:
    z_list.append(feature)

for zz in z_list :
  if zz not in y_list:
    sdvDF[zz] = preprocessing.scale(sdvDF[zz])

#====================== Dummy Varaibles =====================#
#encoder = ce.OneHotEncoder(use_cat_names=True)

for feature in features:
  if (sdvDF.dtypes[feature] in dv_dataType) and (feature not in y_list):
    dv_list.append(feature)

print(dv_list)

for dv in dv_list:
  dummy = pd.get_dummies(sdvDF[dv], drop_first=True)
  
  sdvDF = pd.concat([sdvDF,dummy], axis=1)

df1 = sdvDF.drop((dv_list), axis=1)

df1.reset_index()
```

## Common Utilities

```python
# Common Utilities
import matplotlib.pyplot as plt
import seaborn as sns

def plotVisualization(pred,y_test):
  plt.figure(figsize=(10,6))
  sns.regplot(pred,y_test,scatter_kws={'color':'red','edgecolor':'blue','linewidth':'0.7'},line_kws={'color':'black','alpha':0.5})
  plt.xlabel('Predictions')
  plt.ylabel('Player Valye')
  plt.title("Linear Prediction of Player Value")
  plt.show()
```

## Liner Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LassoCV, Lasso,LassoLarsCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

def getModel(model):
  if model == 'Linear':
    return LinearRegression()
  if model == 'Decision':
    return DecisionTreeRegressor()
  if model == 'Lasso':
    eps = 1e-3
    return LassoCV(eps=eps,cv=10) 

#Spliting data into test and train 
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state = 1234)

models = ['Linear','Decision','Lasso']

for model in models:
  print("\n=============== Metrics to Evaluate Model for {} ========================".format(model))
  mdl = getModel(model)
  mdl.fit(X_train, y_train)
  pred = mdl.predict(X_test)

  print("RMSE: %.2f" %(mean_squared_error(y_test, pred)) ** 0.5)
  print("Variance Score R-Squared: %.2f" %r2_score(y_test,pred))
  print("MSE θ zero Intercept: %.2f" %mean_squared_error(y_test, pred))

  plotVisualization(pred,y_test)
```

```python
# Splitting Database on player positions
def getClass(each):
  yDF = []

  for position in df['Player_Position']:
    if position == each:
      yValue = 1
      yDF.append(yValue)
    else:
      yValue = 0
      yDF.append(yValue)
    
  y = yDF
  X = df1.drop(['NRM_Value_Mil','Player_Position','Cannot Find'],axis=1)

  return X, y
```

```python
positions = ['AM', 'DF', 'DM', 'MF', 'ST']
#positions = ['ATT']

for each in positions:
  print("\n======= Model Performance for player in {} Positions ===========".format(each))
  X,y = getClass(each)


  #Spliting data into test and train 
  X_train , X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state = 1234)

  clf = svm.OneClassSVM(kernel='rbf', gamma='auto')

  clf.fit(X_train)

  y_pred = clf.predict(X_test)

  # Change the anomalies' values to make it consistent with the true values
  prediction = [1 if i==-1 else 0 for i in y_pred]

  # Check the model performance
  print(classification_report(y_test, prediction))
```
